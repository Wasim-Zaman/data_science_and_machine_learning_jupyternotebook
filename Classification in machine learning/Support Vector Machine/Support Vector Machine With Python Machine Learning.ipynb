# Welcome to SVM

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Grab the in-built breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
# let's see the keys of the cancer variable
cancer.keys()
```

## Creating dataset 

```python
dataset = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names'])
```

```python
dataset.head()
```

## Data preprocessing

```python
X = dataset
y = cancer['target']
```

```python
# see the X
X
```

```python
# See the y
y
```

## Splitting the data into train and test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=36)
```

## Creating a model classifier

```python
from sklearn.svm import SVC

# create a classifier
classifier = SVC()

# Fitting the classifier with training data
classifier.fit(X_train, y_train)
```

```python
# predicting the model
y_preds = classifier.predict(X_test)

# showing the predictions
y_preds
```

```python
## Evaluating the model
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_preds))
print('\n')
print(classification_report(y_test, y_preds))
```

our model needs to have its parameter adjusted and it may help to normalize the data as well when you passing it to support vector machine.
What we can do here is to search for best parameters using grid search.

Grid Search:- 
It allows you to find the right parameters such as what c or gamma values to use indide the support vector machine which can be really very difficult task.
**One thing to note** is that the parameter c and gamma can be adjusted via grid search

## Adjusting Parameters
c controls the cost of misclassification on the training data which means that a large c value gives you low bias and high varience whereas low c value gives you high verience and low bias this phenomena is called bias and varience trade off.

```python
from sklearn.model_selection import GridSearchCV

# set the param_grid so that we can feed it to the grid search
param_grid = {'C': [0.01, 0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]}

# setting up the grid
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, verbose=3)
```

```python
# fitting the grid with train data
grid.fit(X_train, y_train)
```

Once we fit our estimator using grid search, we can find best parameter setting as well as the best estimator so that later on if you can set this without using GridSearch

```python
# checking the best parameters 
best_parameters = grid.best_params_

print (best_parameters)
```

```python
# checking out the best estimator for fitting the classifier on
best_estimator = grid.best_estimator_

print (best_estimator)
```

```python
# Fitting the estimator

grid.fit(X_train, y_train)
```

```python
# checking out the predictions
y_preds = grid.predict(X_test)
```

```python
# Evaluating the grid estimator
print(confusion_matrix(y_test, y_preds))

print('\n')

print(classification_report(y_test, y_preds))
```

Hence we can see that after adjusting estimator and parameters via GridSearch, our model (classifier) is now predicting well
